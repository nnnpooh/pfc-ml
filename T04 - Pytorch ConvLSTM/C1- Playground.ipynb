```python
import torch.nn as nn
import torch
from torchviz import make_dot, make_dot_from_trace
```

```python
num_features = 10
filter_size = 3
batch_size = 10
shape = (5, 5)  # H, W
inp_chans = 3
nlayers = 2
seq_len = 4

```

```python
padding = int((filter_size - 1) / 2)
print(padding)

```

```python
conv = nn.Conv2d(
    in_channels=inp_chans + num_features,
    out_channels=4 * num_features,
    kernel_size=filter_size,
    stride=1,
    padding=padding,
)

```

```python
input = torch.randn(1, inp_chans + num_features, shape[0], shape[1])
print(input.size())
```

```python
output = conv(input)
print(output.size())
```

```python
make_dot(conv(input))
```
