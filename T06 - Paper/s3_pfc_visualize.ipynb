```python
import matplotlib.pyplot as plt
import os
import numpy as np
from utils import read_data
```

```python
cwd = os.getcwd()
dm = {
    "pfc1": {
        "sourceName": os.path.join("o2_base_train"),
        "folderName": "2024-01-02-13-00-41",
    },
    "pfc8": {
        "sourceName": os.path.join("o3_pfc_train", "p8"),
        "folderName": "2024-01-02-13-02-55",
    },
    "pfc16": {
        "sourceName": os.path.join("o3_pfc_train", "p16"),
        "folderName": "2024-01-02-13-01-02",
    },
}

dataName = "pfc16"

folderName = dm[dataName]["folderName"]
sourceName = dm[dataName]["sourceName"]

folderPath = os.path.join(os.getcwd(), sourceName, folderName)

data_pickle, data_json = read_data(folderPath)
data = data_pickle["data"]
tArray = data_pickle["tArray"]
nx = data_json["s1"]["n"]
dx = data_json["s1"]["dx"]
L = data_json["s1"]["L"]
tStart = data_json["s1"]["tStart"]
tEnd = data_json["s1"]["tEnd"]
dt = data_json["s1"]["dt"]
mTotal = data_json["s1"]["mTotal"]
```

```python
filenamePrefix = "profile"
nFig = 10
minV = data[-1, :].min()
maxV = data[-1, :].max()
minV = minV - abs(maxV) * 0.2
maxV = maxV + abs(maxV) * 0.2


idxTSample = np.floor(np.linspace(0, tArray.shape[0] - 1, nFig)).astype(int)
xp = np.arange(0, L, dx)
for t in idxTSample:
    tTime = tArray[t]
    fig, ax = plt.subplots(figsize=(8, 3))
    ax.plot(xp, data[t, :], "*--")
    ax.set_title(f"Ground Truth, Time={tTime}")
    ax.set_ylim([minV, maxV])
    # plt.grid(True)
    tTimeFmt = f"{tTime:2.1f}".replace(".", "_")
    filename = f"{filenamePrefix}-{dataName}-t{tTimeFmt}.png"
    # filePath = os.path.join(os.getcwd(), 'o5_visualize', filename)
    # filepath = os.path.join(os.getcwd(), "o5_visualize", filename)
    # plt.savefig(filepath, dpi=300, bbox_inches="tight")
```
