```python
import sys, os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from utilsVisualize import getModelDataNum
import numpy as np
```

```python
mode = "TEST"
modelNames = ["p1"]
visualizeMode = "FE"
cutoffRow = None
filenamePrefix = "fe-time"
```

```python
for modelName in modelNames:
    df, dfLong, dfWide, dfS7 = getModelDataNum(
        modelName, mode, visualizeMode=visualizeMode, cutoffRow=cutoffRow
    )
    print(modelName, df.shape)
```

```python
folders = df.columns.values

fig, ax = plt.subplots()
for idx, folder in enumerate(folders[:5]):
    yp = df[folder]
    xp = df.index.values
    ax.plot(xp, yp, linewidth=3, label=f"p1: Run {idx + 1}")
    ax.set_xlabel("Time")
    ax.set_ylabel("Free Energy")
    
plt.legend()
filename = f'{filenamePrefix}-{modelName}.png'
filePath = os.path.join(os.getcwd(), 'o8_visualize', filename)
plt.savefig(filePath, dpi=300, bbox_inches="tight")
```
