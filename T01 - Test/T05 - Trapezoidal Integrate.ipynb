# Test Trapezoidal Integration

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example from documentation
theta = np.linspace(0, 2 * np.pi, num=1000, endpoint=True)
np.trapz(np.cos(theta), x=np.sin(theta))
```

```python
n = 128 # Number of grid
L = 8 * np.pi # Length of the domain
dx = L / n  # Grid spacing

x = np.linspace(0, L-dx, num=n)
phi = np.sin(x)

plt.plot(x, phi, "-*")
```

```python
phi
```

```python
phi[:1]
```

```python
xe = np.linspace(0, L, num=n+1)
ye = np.concatenate((phi, phi[:1])) 

plt.plot(xe, ye, "-*")
```

```python
np.trapz(xe, ye)
```
