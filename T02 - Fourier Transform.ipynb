# Test Fourier Transform

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Pi = np.pi
n = 100
L = 2 * np.pi
dx = L/ n
```

```python
# x and y
x = np.linspace(0, 2 * Pi - dx, n)
y = np.sin(2 * Pi * x / L)
plt.figure()
plt.plot(x, y, "*")
```

```python
# Calculate laplacian of y
y_hat = np.fft.fft(y)
k = 2 * Pi / L * np.concatenate((np.arange(0, n / 2), np.arange(-n / 2, 0)))  # or np.fft.fftfreq(100, d=dx/L)
k2 = np.power(k, 2)
y_hat2 = np.multiply(y_hat, -np.absolute(k2))
lap_y = np.fft.ifft(y_hat2)
plt.figure()
plt.plot(x, np.real(lap_y), "*")

# Error
plt.figure()
plt.plot(x,np.real(lap_y+y))
```
